To import the files, it needs to access your Google Drive.

It will then ask you to click a link to authenticate, from where you get a code and paste it here.

This will kind of give the code the power to access all your files in your drive, so it can get the data from .oszs.

```
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
from google.colab import auth
from oauth2client.client import GoogleCredentials

# 1. Authenticate and create the PyDrive client.
auth.authenticate_user()
gauth = GoogleAuth()
gauth.credentials = GoogleCredentials.get_application_default()
drive = GoogleDrive(gauth)
```

Then, import some maps from your Google Drive.

First, make a folder containing those .osz files; then replace the folder ID with your folder (it is obvious when you open it) and run this grid.

```
import subprocess, os

# your google drive folder containing oszs
folder_id = '1UAtsBa655BpRSmm553-FabRIcaTeD668';

# make a maps folder
# if it already exists just ignore the error!
!mkdir maps
!mkdir mapdata

# get all files from that folder.
# and unzip them because they are oszs!
top_list = drive.ListFile({'q': "'{}' in parents and trashed=false".format(folder_id)}).GetList()
for file in top_list:
  file.GetContentFile("maps/" + file['title']);
  subprocess.call(["unzip", "maps/" + file['title'], "-d", "maps/" + os.path.splitext(file['title'])[0] + "/"]);
  
!ls maps/
```
